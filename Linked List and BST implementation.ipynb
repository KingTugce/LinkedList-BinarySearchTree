```python
class Node():
    def __init__(self, value):
        self.value = value  # input value
        self.next = None  # this is the stored link

class Linked():
    def __init__(self):
        self.groundzero = None   # set the starter list
    
    def iterate(self):
        if self.groundzero == None:            # if no root node then cannot iterate
            print("This linked list is empty!")
            return
        else:
            container = self.groundzero
            while container.next:        # set a flag to stop when list is cycled
                print(container.value)
                container = container.next  # set a counter
            print(container.value)   #catch the last value
            
    def append(self, value):
        if self.groundzero == None:         # Fixed logic to only require an append and not a pushon to start list
            new_node = Node(value)                 # Would not correctly start the first link in the list otherwise
            new_node.next = self.groundzero
            self.groundzero = new_node    # calls node to add to an empty list. confusing because of "groundzero"
        else:
            container = self.groundzero
            while container.next != None:       #Set a stop flag
                container = container.next  # Cycle all values since we REDEIFNE container as THE NEXT ONE
            container.next = Node(value)
            
mygrades = Linked()
mygrades.append('A')
mygrades.append('B')
mygrades.append('C')
mygrades.iterate()
```

```python
class BST():
    def __init__(self,value):
        self.value = value # Current Value
        self.left = None
        self.right = None
        
    def insert(self,value):
        if value < self.value:
            if self.left is None:
                self.left = BST(value)
            else:
                self.left.insert(value)
        else:
            if self.right is None:
                self.right = BST(value)
            else:
                self.right.insert(value)
        return self
    
    def contains(self,value):
        if value < self.value: # if passed value is less than current value
            if self.left is None:
                return False
            else:
                return self.left.contains(value)
        elif value > self.value:
            if self.right is None:
                return False
            else:
                return self.right.contains(value)
        else:
            return True
        
    def getMinValue(self):
        if self.left is None:
            return self.value
        else:
            return self.left.getMinValue()
    
    def getMaxValue(self):
        if self.right is None:
            return self.value
        else:
            return self.right.getMaxValue()
        
    def remove(self,value,parent=None):
        if value < self.value:
            if self.left is not None:
                self.left.remove(value,self)
        elif value > self.value:
            if self.right is not None:
                self.right.remove(value,self)
        else:
            if self.left is not None and self.right is not None:
                self.value = self.right.getMinValue()
                self.right.remove(self.value,self)
            elif parent is None:
                if self.left is not None:
                    self.value = self.left.value
                    self.right = self.left.right
                    self.left = self.left.left
                elif self.right is not None:
                    self.value = self.right.value
                    self.left = self.right.left
                    self.right = self.right.right
                else:
                    self.value = None
            elif parent.left == self:
                parent.left = self.left if self.left is not None else self.right
            elif parent.right == self:
                parent.right = self.left if self.left is not None else self.right
        return self
    
bst_example = BST(89)
bst_example.insert(60)
bst_example.insert(61)

print(bst_example.getMaxValue())
print(bst_example.contains(88))
bst_example.remove(90)
print(bst_example.contains(90))
```
